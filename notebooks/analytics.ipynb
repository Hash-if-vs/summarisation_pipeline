```python
import pandas as pd
df= pd.read_csv('../results/summarization_scores.csv')
df
```

```python


# Group by dialogue to avoid repetition
for i, (dialogue, group) in enumerate(df.groupby('dialogue'), 1):
    print(f"Example {i}")
    print(f"Dialogue:\n{dialogue.strip()}\n")
    
    reference = group['reference_summary'].iloc[0]
    print(f"Reference Summary:\n{reference.strip()}\n")
    
    print("Generated Summaries:")
    for _, row in group.iterrows():
        print(f"  Model: {row['model']}")
        print(f"  Generated: {row['generated_summary'].strip()}\n")
    
    print("-" * 100)
```
